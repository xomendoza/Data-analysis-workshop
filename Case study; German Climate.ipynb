## 1. Prepocessing the German climate data

```python
# Importing pandas library

import pandas as pd
```

```python
# Reading the files

precipitation_data = pd.read_csv('c:01001_195101_201412.txt', delimiter = ';')

vapor_data = pd.read_csv('c:98836_196801_201712.txt', delimiter = ';')

sunshine_data = pd.read_csv('c:98836_197803_201612.txt', delimiter = ';')
```

```python
# Displaying the first five rows of precipitation data
precipitation_data.head()
```

```python
# Displaying the first five rows of vapor data
vapor_data.head()
```

```python
# Displaying the first five rows of sunshine data
sunshine_data.head()
```

```python
# Converting dataset from wide format to long format for precipitation data

precipitation_data = pd.melt(precipitation_data,\
                             id_vars=['Jahr'],\
                             value_vars = ['Jan', 'Feb', 'Mrz', 'Apr', 'Mai', 'Jun', 'Jul', 'Aug', 'Sep', 'Okt', 'Nov', 'Dez'])

precipitation_data.head()

```

```python
# Renaming the columns names of precipitation dataframe

precipitation_data = precipitation_data.rename(columns={ 'value': 'Precipitation'})

precipitation_data.head()
```

```python
# Converting dataset from wide format to long format for vapor data

vapor_data = pd.melt(vapor_data,
                    id_vars = ['Jahr'], 
                    value_vars = ['Jan', 'Feb', 'Mrz', 'Apr', 'Mai', 'Jun', 'Jul', 'Aug', 'Sep', 'Okt', 'Nov', 'Dez'])


# Renaming the column names of vapor data

vapor_data = vapor_data.rename(columns={'value':'Vapor'})

vapor_data.head()
```

```python
# Converting dataset from wide format to long format for sunshine data

sunshine_data = pd.melt(sunshine_data,
                    id_vars = ['Jahr'], 
                    value_vars = ['Jan', 'Feb', 'Mrz', 'Apr', 'Mai', 'Jun', 'Jul', 'Aug', 'Sep', 'Okt', 'Nov', 'Dez'])


# Renaming the column names of sunshine data

sunshine_data = sunshine_data.rename(columns={'value':'Sunshine'})

sunshine_data.head()
```

## 2. Merging DataFrames and renaming variables

```python
# Merging the first two datasets together 

conDf = pd.merge(precipitation_data, vapor_data,\
                how='inner',\
                on=['Jahr','variable'])

conDf.head()
```

```python
# Merging the third dataset

conDf = pd.merge(conDf, sunshine_data,\
                how = 'inner', on = ['Jahr', 'variable'])

conDf.head()
```

```python
# Converting the year and month variables from German to English format

months = {'mrz':'Mar', 'Mai':'May','Okt':'Oct', 'Dez':'Dec'}

conDf['variable'] = \
conDf['variable'].map(months).fillna(conDf['variable'])
conDf
```

```python
# Renaming the Jahr variable to Year

conDf = conDf.rename(columns={'Jahr':'Year'})

conDf
```

## 3. Data interpolation and answering questions after data preprocessing 

```python
# Introducing a new column called months to set months in numerical order instead of alphabetical 

conDf['Months'] = conDf['variable']
conDf.head()
```

```python
# Creating a dictionary to map the month to the corresponding numerical order 

monthsMap = {'Jan':1, 'Feb':2, 'Mrz':3, 'Apr':4, 'May':5, 'Jun':6,\
             'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12}
```

```python
# Using the dictionary  and maping the new column to their corresponding numerical equivalents 
conDf['Months'] = conDf['Months'].map(monthsMap).fillna(conDf['Months'])
conDf
```

```python
# Adding more code for sorting the months 

newCondf = conDf.sort_values(by=['Year','Months'])
newCondf
```

```python
# Imputing missing values using the df.interpolate() method, assuming the there is a linear trend for different variables

cleanDf1 = newCondf.interpolate(method='linear',\
                                limit_direction='both')

cleanDf1
```

##### Finding the average vapor pressure for the month of January across all the years 

```python
# Grouping the dataset on the variable column and then do a mean aggregation for the vapor_pressure column and filter january

Q1 = cleanDf1.groupby(['variable'])['Vapor'].agg('mean').loc['Jan']
Q1
```
